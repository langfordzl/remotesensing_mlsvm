# Load Data

```python
import sys
import numpy as np

sys.path.insert(0, '/home/lql/Desktop/rs_mlsvm/repo/petsc/lib/petsc/bin')
import PetscBinaryIO
```

```python
# data path

p = '/home/lql/Desktop/rs_mlsvm/repo/mlsvm/src/datasets/'


def load_data(p, x, y):
    xf=p+x
    yf=p+y
    io = PetscBinaryIO.PetscBinaryIO()
    sparse_data= io.readBinaryFile(xf, mattype='scipy.sparse')
    xdata = sparse_data[0].toarray()

    # convert data to array
    io = PetscBinaryIO.PetscBinaryIO()
    sparse_data= io.readBinaryFile(yf, mattype='scipy.sparse')
    ydata = sparse_data[0]

    print(xdata.shape)
    print(ydata.shape)

    yd = np.asarray(ydata)

    (unique, counts) = np.unique(yd, return_counts=True)
    print(np.asarray((unique, counts)).T)
    
    return xdata, yd




xp = 'ak03_fire_train2_zsc_data.dat'
yp = 'ak03_fire_train2_label.dat'

xt, yt = load_data(p, xp, yp)
```

```python
xp = 'val_data_zsc_data.dat'
yp = 'val_data_label.dat'

xv, yv = load_data(p, xp, yp)
```

# Load MLSVM Model

```python
from libsvm.svm import *
from libsvm.svmutil import *
import os
from sklearn.metrics import classification_report

def loop_directory(rootdir, extensions, y1, x2):
    report = {}
    for subdir, dirs, files in os.walk(rootdir):
        for file in files:
            ext = os.path.splitext(file)[-1].lower()
            if ext in extensions:
                print (os.path.join(subdir, file))
                filename = os.path.join(subdir, file)
                m = svm_load_model(filename)
                pred_lbl, pred_acc, pred_val = svm_predict(y1,x2,m)
                print(classification_report(y1, pred_lbl))
                creport = classification_report(y1, pred_lbl)
                report[filename] = creport
    return report
```

```python
rootdir = '/home/lql/Desktop/rs_mlsvm/repo/mlsvm/src/svm_models'
extensions = ('.svmmodel')

report = loop_directory(rootdir, extensions, yt, xt)
keys_train = list(report.keys())
```

```python
report = loop_directory(rootdir, extensions, yv, xv)
keys_val = list(report.keys())
```
